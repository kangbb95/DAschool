```python
# 데이터 타입들 중 컬럼들이 서로 상관이 있거나 반대 성향이 있으면 서로 비교
pos = kr['확진자']
neg = kr['결과 음성']
```

```python
plt.plot(pos, 'red')
plt.plot(neg, 'blue')

plt.xticks(rotation = 90)
plt.legend(['양성','음성'])
plt.show()
```

```python
# 두 데이터(컬럼) 합쳐서 하나의 결과를 도출 -> 컬럼 선택
# 확진자 수 중 사망 -> 사망율
# 환진자 수: pos
pos
death = kr['사망']
# death/pos * 100 -> 비율
```

```python
d_p = round((death / pos) * 100, 2)
d_p.head()
```

```python
# 완치율
# 확진자 수
# 격리해제 수
pos
rec = kr['격리해제']

# rec/pos * 100 -> 비율
```

```python
r_p = round((rec / pos) * 100, 2)
r_p.head()
```

```python
# 치명율, 완치율

plt.plot(d_p)
plt.plot(r_p)
plt.xticks(rotation = 90)
plt.show()
```

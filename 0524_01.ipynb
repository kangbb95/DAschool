### 데이터프레임 병합

```python
data = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]
    }, index=['a', 'b', 'c', 'd']
)

data2 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]},
    index=['e', 'f', 'g', 'h']
    )
```

```python
add_data = pd.concat([data2, data]) # 두 개의 데이터를 합쳐줌
```

```python
add_data
```

```python
add_data = pd.concat([data, data2], ignore_index = True) # 두 개의 데이터를 합쳐줌 # ignore_index = True -> 인덱스 재배열
```

```python
add_data
```

```python
data = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]
    }
)

data2 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]}
    )
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                  'A': ['A0', 'A4', 'A2', 'A3'],
            'B': ['B0', 'B4', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                  'C': ['C0', 'C1', 'C2', 'C3'],
                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
print(left)
print(right)
```

```python
mi = pd.merge(left, right)
```

```python
print(mi)
```

```python
mo = pd.merge(left, right, how = 'outer')
```

```python
print(mo)
```

```python
m_key = pd.merge(left, right, on='key')
print(m_key)
```

```python
m_l = pd.merge(left, right, on='key', how='left')
print(m_l)
```

```python
m_r = pd.merge(left, right, on ='key', how='right')
print(m_r)
```
